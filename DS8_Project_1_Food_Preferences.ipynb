You have recently purchased restaurant space in the UCLA food court and are trying to understand the food preferences of college students to decide what kind of restaurant you'll create. Your co-owner conducted a survey and has compiled a CSV file with the results for you to analyze.


Begin by importing the dataset (https://static.junilearning.com/python_data_science/food.csv) as a DataFrame, and then clean the DataFrame by dropping any rows with null values.


```
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```
foodDF = pd.read_csv("https://static.junilearning.com/python_data_science/food.csv")
display(foodDF)
```

```
foodDF_clean = foodDF.dropna()
display(foodDF_clean)
```

Now let's determine which type of food is the favorite! Students in the survey wrote in their favorite cuisine (`fav_cuisine`) and also chose from a list of 9 choices (`fav_cusine_encoded`). The `fav_cusine_encoded` will be easier for us to analyze, so that's what we'll be using for our plots.

Create a bar graph displaying the amount of choices for each category.<br>
Based on the results, what type of restaurant would you suggest creating?

```
decoding = {0:"None", 1:"Italian", 2:"Mexican", 3:"Arabic", 4:"Asian", 5:"American", 6:"African", 7:"Jamaican", 8:"Indian"}
```

```
foodDF_clean["fav_cuisine_decoded"]=foodDF_clean["fav_cuisine_coded"].map(decoding)    
display(foodDF_clean)
```

```
sns.countplot(x="fav_cuisine_decoded",data=foodDF_clean).set(title="Cuisine Ranked By Popularity")
#The best type of restaurant would be Italian.
```

Your co-owner is impressed with your work! However, they want to make sure that your restaurant will have broad appeal. How would people feel about your restaurant even if wasn't their top choice? Would they still order there occasionally?

The `italian_food` column in our DataFrame records how likely a student is to eat Italian food on a scale of 1 to 5:

1 - very unlikely 

2 - unlikely 

3 - neutral 

4 - likely 

5 - very likely 

Create a bar chart plotting the `italian_food` column. Then create a box plot plotting `italian_food` against `fav_cuisine_encoded`. 

```
sns.countplot(x="italian_food",data=foodDF_clean).set(title="How Likely To Eat Italian Food")
```

```
sns.boxplot(y="fav_cuisine_decoded",x="italian_food",data=foodDF_clean).set(title="How Likely To Eat Italian Food")
```

What conclusions can we draw from these plots? 

Student Analysis:
The populations that voted for top choices (Asian, American, Mexican, and Indian) would still be happy if presented with Italian food. Based on these findings, I would still recommend creating an Italian restaurant because it seems to have wide appeal.
